```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder
import re
```

```python
df = pd.read_csv("student.csv")
```

เป็นคำสั่งใน pandas ที่ใช้ อ่านไฟล์ CSV แล้วเก็บข้อมูลไว้ในตัวแปร df ในรูปแบบ DataFrame

```python
df
```

df โชว์ตารางข้อมูล (DataFrame) ทั้งหมด ของตัวแปร df ให้ดูแบบสวยงามในรูปแบบตาราง

```python
df_clean = df.copy()
```

ใช้เพื่อ สร้างสำเนาของ DataFrame df ขึ้นมาใหม่อีกตัวหนึ่งชื่อ df_clean หรือพูดง่ายๆคือเป็นการก็อปปี้ข้อมูลไว้อีกชุดแยกต่างหาก

```python
df_clean.columns = df_clean.columns.str.strip()
```

df_clean.columns = df_clean.columns.str.strip() เป็นการลบช่องว่าง (space) ด้านหน้าและด้านหลังของชื่อคอลัมน์ใน DataFrame ทำให้คอลัมน์สะอาดขึ้น

```python
df_clean.loc[df_clean['GPA'] < 0, 'GPA'] = np.nan  # Remove negative GPA
df_clean.loc[df_clean['GPA'] > 4.0, 'GPA'] = 4.0   # Cap at 4.0
```

โค้ดนี้เป็นการจัดการค่าผิดปกติในคอลัมน์ 'GPA' ของ DataFrame df_clean โดยใช้ .loc เพื่อเลือกและแก้ไขข้อมูลตามเงื่อนไข หรือทำความสะอาดข้อมูล GPA โดยลบค่าติดลบและปรับค่าที่เกินขอบเขตให้เหมาะสม

```python
df_clean.loc[df_clean['อายุ'] > 30, 'อายุ'] = np.nan  # Remove unrealistic age (150)
df_clean.loc[df_clean['อายุ'] < 17, 'อายุ'] = np.nan
```

โค้ดนี้ใช้ ลบหรือแทนค่าข้อมูลอายุที่อยู่นอกช่วง 17-30 ปีด้วยค่าว่าง (NaN) เพื่อให้ข้อมูลสะอาดและเหมาะสมกับการวิเคราะห์ต่อไป

```python
df_clean.loc[df_clean['ส่วนสูง_cm'] < 140, 'ส่วนสูง_cm'] = np.nan
df_clean.loc[df_clean['ส่วนสูง_cm'] > 200, 'ส่วนสูง_cm'] = np.nan
```

 โค้ดนี้ช่วย กรองและลบข้อมูลส่วนสูงที่อยู่นอกช่วง 140–200 ซม. เพื่อให้ข้อมูลสะอาดและเชื่อถือได้มากขึ้น ก่อนนำไปวิเคราะห์ต่อ

```python
df_clean.loc[df_clean['น้ำหนัก_kg'] < 35, 'น้ำหนัก_kg'] = np.nan
df_clean.loc[df_clean['น้ำหนัก_kg'] > 120, 'น้ำหนัก_kg'] = np.nan
```

โค้ดนี้เป็นการทำความสะอาดข้อมูลในคอลัมน์ 'น้ำหนัก_kg' ของ DataFrame df_clean โดยแทนค่าที่ผิดปกติด้วย NaN เพื่อบ่งบอกว่าข้อมูลนี้ไม่มีค่า

```python
df_clean.loc[df_clean['ชั่วโมงเรียน_ต่อสัปดาห์'] < 5, 'ชั่วโมงเรียน_ต่อสัปดาห์'] = np.nan
df_clean.loc[df_clean['ชั่วโมงเรียน_ต่อสัปดาห์'] > 40, 'ชั่วโมงเรียน_ต่อสัปดาห์'] = np.nan
```

โค้ดนี้เป็นการทำความสะอาดข้อมูลในคอลัมน์ 'ชั่วโมงเรียน_ต่อสัปดาห์' ของ DataFrame df_clean คือเลือกแถวที่ค่าชั่วโมงเรียนต่อต่ำกว่า 5 ชั่วโมงต่อสัปดาห์ ซึ่งอาจน้อยเกินไปหรือผิดพลาด
แล้วแทนค่าดังกล่าวด้วย NaN เพื่อบ่งบอกว่าข้อมูลขาดหายหรือไม่ถูกต้อง

```python
gender_mapping = {
        'ชาย': 'ชาย',
        'หญิง': 'หญิง',
        'ญ': 'หญิง',
        'ช': 'ชาย',
        'M': 'ชาย',
        'F': 'หญิง',
        'Male': 'ชาย',
        'Female': 'หญิง'
    }

df_clean['เพศ'] = df_clean['เพศ'].map(gender_mapping)
```

โค้ดนี้เป็นการ ปรับค่าข้อมูลในคอลัมน์ 'เพศ' ของ DataFrame df_clean ให้เป็นรูปแบบที่สอดคล้องกัน ใช้ .map() แทนค่าทุกค่าในคอลัมน์เพศ ด้วยค่าที่ตรงกับ gender_mapping

```python
faculty_mapping = {
        'วิทยาศาสตร์': 'Science',
        'วิทย์': 'Science',
        'Science': 'Science',
        'Sci': 'Science',
        'คณะวิทยาศาสตร์': 'Science',
        'วิศวะ': 'Engineering',
        'วิศวกรรม': 'Engineering',
        'วิศวฯ': 'Engineering',
        'วิศวกรรมศาสตร์': 'Engineering',
        'Engineering': 'Engineering',
        'ศิลปศาสตร์': 'Liberal Arts',
        'ศิลป์': 'Liberal Arts',
        'Arts': 'Liberal Arts',
        'Liberal Arts': 'Liberal Arts'
    }

df_clean['คณะ'] = df_clean['คณะ'].map(faculty_mapping)
```

ค้ดนี้ช่วยให้ข้อมูลคณะเป็นระเบียบและง่ายต่อการวิเคราะห์ด้วยการรวมชื่อคณะที่เหมือนกันแต่เขียนต่างรูปแบบให้อยู่ในรูปแบบเดียวกัน เช่น คำที่เกี่ยวกับวิทยาศาสตร์ เช่น วิทยาศาสตร์ เป็น วิทย์, Science เป็น Sci

```python
for col in ['เพศ', 'คณะ']:
    df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)
```

โค้ดนี้เป็นการเติมค่าที่หายไป (missing values) ในคอลัมน์ เพศ และ คณะ ของ DataFrame df_clean ด้วยค่าที่พบมากที่สุดในแต่ละคอลัมน์นั้น คือ for col in ['เพศ', 'คณะ']: ทำงานวนลูปทีละคอลัมน์ คือ 'เพศ' และ 'คณะ'

```python
for col in [ 'อายุ', 'GPA', 'ส่วนสูง_cm', 'น้ำหนัก_kg', 'ชั่วโมงเรียน_ต่อสัปดาห์']:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mean())
df_clean['อายุ'] = df_clean['อายุ'].astype(np.double)
```

ส่วนแรกทำลูปผ่านคอลัมน์ที่ระบุ ได้แก่ 'อายุ', 'GPA', 'ส่วนสูง', 'น้ำหนัก' และ 'ชั่วโมงเรียนต่อสัปดาห์' ในแต่ละคอลัมน์ จะเติมค่าที่หายไปด้วยค่าเฉลี่ย ของคอลัมน์นั้น ช่วยให้ข้อมูลสมบูรณ์มากขึ้น โดยไม่ใช้ค่า NaN ในการวิเคราะห์
ส่วนที่สอง คือ แปลงชนิดข้อมูลของคอลัมน์ 'อายุ' ให้เป็นตัวเลขทศนิยม เพื่อให้แน่ใจว่าค่าในคอลัมน์อยู่ในรูปแบบตัวเลขสำหรับการคำนวณ
